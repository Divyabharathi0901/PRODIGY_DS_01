Data Visualisation

```python
plt.figure(figsize=(10, 5))
plt.bar(population_data.index, population_data['2022'], color='blue', label='2022')
plt.bar(population_data.index, population_data['1961'], color='red', label='1961', alpha=0.7)
plt.title('World Population Growth Rate 1961 to 2022')
plt.xlabel('Country')
plt.ylabel('Population Growth (%)')
plt.legend()
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility if there are many countries/regions
plt.tight_layout()
plt.show()
```

```python
# Plotting the bar diagram
plt.figure(figsize=(25, 8))
sns.barplot(x='country_name', y='1991', data=population_data, color='skyblue')
plt.title('World Population Growth Rate')
plt.xlabel('Country Name')
plt.ylabel('Population Growth Rate in 1991 (%)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
# Selecting the column containing population growth rates
growth_rates = population_data.drop(columns=['country_code', 'country_name']).values.flatten()

# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(growth_rates, bins=30, kde=True, color='blue')
plt.title('Distribution of Population Growth Rates')
plt.xlabel('Population Growth (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='2021', y='2022', data=population_data)
plt.title('Scatter Plot: Population Growth Rate (2021) vs. Population Growth Rate (2022)')
plt.xlabel('Population Growth Rate (2021)')
plt.ylabel('Population Growth Rate (2022)')
plt.tight_layout()
plt.show()
```

```python
# List of columns representing the years from 1961 to 2022
years = list(map(str, range(1961, 2023)))  # Create a list of years as strings

# Check if the required columns exist
missing_years = [year for year in years if year not in population_data.columns]
if missing_years:
    print(f"Missing columns: {missing_years}")
else:
    # Summing the population growth percentages across the years for each country
    population_data['total_growth'] = population_data[years].sum(axis=1)

    # Grouping by country and summing the total growth, then sorting and selecting the top 10 countries
    pop_2022 = population_data.groupby('country_name')[['total_growth']].sum().sort_values('total_growth', ascending=False).round(2).head(10)

    # Resetting the index to make 'country_name' a column again
    pop_2022.reset_index(inplace=True)

    # Creating a bar plot with Plotly
    fig = px.bar(pop_2022, x='country_name', y='total_growth',
                 title='Top 10 Countries by Population Growth (%)',
                 template='seaborn', color='country_name', text='total_growth')

    # Updating layout for better readability
    fig.update_layout(xaxis_title='Country',
                      yaxis_title='Population Growth (%)',
                      title_x=0.5)  # Center the title

    fig.show()
```

```python
# List of columns representing the years from 1961 to 2022
# Using histogram
years = list(map(str, range(1961, 2023)))  # Create a list of years as strings

# Check if the required columns exist
missing_years = [year for year in years if year not in population_data.columns]
if missing_years:
    print(f"Missing columns: {missing_years}")
else:
    # Summing the population growth percentages across the years for each country
    population_data['total_growth'] = population_data[years].sum(axis=1)

    # Plotting the histogram of total population growth
    plt.figure(figsize=(10, 6))
    plt.hist(population_data['total_growth'], bins=30, edgecolor='black')
    plt.title('Histogram of Total Population Growth (1961-2022)')
    plt.xlabel('Total Population Growth (%)')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Check if the required columns exist
# Using scatter plot
if '2021' in population_data.columns and '2022' in population_data.columns:
    # Plotting the scatter plot for population growth rates of 2021 vs. 2022
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='2021', y='2022', data=population_data)
    plt.title('Scatter Plot: Population Growth Rate (2021) vs. Population Growth Rate (2022)')
    plt.xlabel('Population Growth Rate (2021)')
    plt.ylabel('Population Growth Rate (2022)')
    plt.tight_layout()
    plt.show()
else:
    print("One or both columns '2021' and '2022' do not exist in the DataFrame.")
```

```python
data = {
    'Country': ['Country1', 'Country2', 'Country3'],
    '1961': [1.2, 2.3, 3.4],
    '1962': [1.5, 2.6, 3.7],
    '2022': [1.8, 2.9, 4.0]
}
population_data = pd.DataFrame(data)

# Transform the DataFrame to have years as a single column
population_data_melted = population_data.melt(id_vars=['Country'], var_name='Year', value_name='Population_Growth')

# Aggregate data by year
population_growth_by_year = population_data_melted.groupby('Year').sum()

# Plotting the pie chart
plt.figure(figsize=(10, 7))
plt.pie(population_growth_by_year['Population_Growth'], labels=population_growth_by_year.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('World Population Growth Rate by Year')
plt.show()
```
